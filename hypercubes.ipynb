```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
E = [(0,1)]
G = [nx.from_edgelist(E)]  # 1 dim
for i in range(2,7):
    G.append(nx.cartesian_product(G[-1],G[0])) 

fig, ax = plt.subplots(2,3, figsize=(10,5))
i = 0
for hypercube in G:
    nx.draw(hypercube, node_size=7, width = 0.8, node_color='darkblue', ax=ax[i//3,i%3])
    ax[i//3,i%3].set_title(f'{i+1}-dimensional hypercube')
    i += 1
fig.savefig('./figs/hypercube_examples.png')
```
